## Problema Estándar 5 (SP5) – Transferencia de Momento de Espín (Spin Transfer Torque)

El **Problema Estándar 5 (SP5)** forma parte de la serie de *Standard Problems in Micromagnetics* propuestos para comparar, validar y analizar simulaciones numéricas de dinámica magnética entre distintos códigos y metodologías.  

Este problema está enfocado en el fenómeno de **Transferencia de Momento de Espín** (*Spin Transfer Torque*, STT), un efecto físico en el que una corriente eléctrica polarizada en espín puede transferir momento angular a la magnetización de un material, provocando su movimiento o inversión (*switching*) sin necesidad de aplicar un campo magnético externo.  

---

###  Historia detrás y relevancia
El STT fue propuesto teóricamente por **Slonczewski (1996)** y **Berger (1996)** como una forma eficiente de controlar la magnetización a escala nanométrica.  
Este mecanismo es clave en el desarrollo de:
- **Memorias MRAM** (Magnetoresistive Random Access Memory).
- Dispositivos de lógica *spintrónica*.
- Control y manipulación de *skyrmions* y vórtices magnéticos.

A diferencia del control tradicional por campos magnéticos, el STT permite una conmutación localizada y con menor consumo energético, ya que actúa directamente a través de la interacción entre los electrones conducidos y la magnetización local.

---

### 2. Geometría y configuración inicial en SP5
En SP5 se considera un **nanodisco** o **nanodot** ferromagnético del orden de 100 nm de diámetro y 10 nm de espesor.  
La configuración inicial es un **vórtice magnético**:  
- La magnetización in-plane circula alrededor del centro.
- En el núcleo del vórtice, \( M_z \) apunta perpendicularmente.

Este estado inicial es estable en ausencia de perturbaciones. Al aplicar una corriente polarizada en espín, la estructura comienza a moverse o distorsionarse dependiendo de la intensidad y características de la corriente.

---

### 3. Dinámica gobernada por la ecuación LLG extendida
La dinámica se describe mediante la **ecuación de Landau–Lifshitz–Gilbert (LLG) extendida** para incluir el torque por corriente de espín:

```{math}
\frac{\partial \mathbf{m}}{\partial t} =
-\gamma \, \mathbf{m} \times \mathbf{H}_{\text{eff}}
+ \alpha \, \mathbf{m} \times \frac{\partial \mathbf{m}}{\partial t}
- (\mathbf{u} \cdot \nabla)\mathbf{m}
+ \xi \, \mathbf{m} \times \left[ (\mathbf{u} \cdot \nabla)\mathbf{m} \right]
```
Donde:
- **Primer término**: precesión de $\mathbf{m} $ alrededor de $\mathbf{H}_{\text{eff}}.$
- **Segundo término**: amortiguamiento de Gilbert.
- **Tercer término**: *torque adiabático* (alineamiento progresivo de los espines de conducción con la magnetización local).
- **Cuarto término**: *torque no adiabático*, proporcional a $\xi$, asociado a desalineamientos y procesos de relajación transversal del espín.

---

### 4. Parámetro de no adiabaticidad $(\xi)$
El parámetro \(\xi\) (también escrito como \(\beta\) en la literatura) controla la contribución no adiabática del torque.  
Su influencia en la dinámica es clave:
- **$\xi \approx 0$** → transporte casi adiabático, el movimiento de texturas es más lento y requiere mayor corriente para producir switching.
- **$\xi$ grande** → mayor velocidad de desplazamiento y switching, pero también mayor posibilidad de distorsión y pérdida de estabilidad.

En SP5 se realizan simulaciones variando $\xi$ para estudiar su efecto sobre:
- La trayectoria de la magnetización promedio $(M_x, M_y).$
- El tiempo y la velocidad de switching.
- La deformación de la textura magnética inicial.

---

### 5. Objetivo del SP5
El objetivo principal es **proporcionar un caso de referencia** para:
- Comparar la implementación del STT en diferentes códigos micromagnéticos.
- Entender la relación entre el torque no adiabático y la eficiencia de la conmutación.
- Evaluar la estabilidad de estructuras topológicas (como vórtices) bajo corrientes de espín.

La simulación parte de un estado de vórtice relajado, y se aplica una corriente polarizada en la dirección **x** durante un tiempo determinado, midiendo la evolución temporal de \((M_x, M_y, M_z)\) y la energía total.

---

### 6. Relevancia actual
El SP5 sigue siendo relevante porque:
- Modela un fenómeno central en *spintrónica*.
- Sirve como banco de pruebas para algoritmos de integración numérica y tratamiento del STT.
- Es base para estudios más complejos, como *skyrmion dynamics*, switching en *synthetic antiferromagnets* o estabilidad de dominios bajo corriente.

````python
def read_mumax3_table(filename):
    """Puts the mumax3 output table in a pandas dataframe"""

    from pandas import read_table
    
    table = read_table(filename)
    table.columns = ' '.join(table.columns).split()[1::2]
    
    return table
````

````python
def read_mumax3_ovffiles(outputdir):
    """Load all ovffiles in outputdir into a dictionary of numpy arrays 
    with the ovffilename (without extension) as key"""
    
    from subprocess import run, PIPE, STDOUT
    from glob import glob
    from os import path
    from numpy import load

    # convert all ovf files in the output directory to numpy files
    p = run(["mumax3-convert","-numpy",outputdir+"/*.ovf"], stdout=PIPE, stderr=STDOUT)
    if p.returncode != 0:
        print(p.stdout.decode('UTF-8'))

    # read the numpy files (the converted ovf files)
    fields = {}
    for npyfile in glob(outputdir+"/*.npy"):
        key = path.splitext(path.basename(npyfile))[0]
        fields[key] = load(npyfile)
    
    return fields
````

````python
def run_mumax3(script, name, verbose=False):
    """ Executes a mumax3 script and convert ovf files to numpy files
    
    Parameters
    ----------
      script:  string containing the mumax3 input script
      name:    name of the simulation (this will be the name of the script and output dir)
      verbose: print stdout of mumax3 when it is finished
    """
    
    from subprocess import run, PIPE, STDOUT
    from os import path

    scriptfile = name + ".txt" 
    outputdir  = name + ".out"

    # write the input script in scriptfile
    with open(scriptfile, 'w' ) as f:
        f.write(script)
    
    # call mumax3 to execute this script
    p = run(["mumax3","-f",scriptfile], stdout=PIPE, stderr=STDOUT)
    if verbose or p.returncode != 0:
        print(p.stdout.decode('UTF-8'))
        
    if path.exists(outputdir + "/table.txt"):
        table = read_mumax3_table(outputdir + "/table.txt")
    else:
        table = None
        
    fields = read_mumax3_ovffiles(outputdir)
    
    return table, fields
````

### Formación de espirales en \(M_x\) y \(M_y\)

Al analizar simultáneamente las componentes \(M_x\) y \(M_y\), observamos patrones de **espiral**  que se originan desde el centro de la textura y se extienden radialmente.

#### Fisicamente :
- **Debido a la configuración inicial tipo vórtice**:  
  En este tipo de estado, la magnetización en el plano (\(M_x, M_y\)) rota progresivamente alrededor del núcleo, formando una estructura circular.
  
- **Relación entre \(M_x\) y \(M_y\)**:  
  Si representamos \((M_x, M_y)\) como un vector 2D en cada punto, su dirección rota angularmente al alejarnos del centro.  
  Esto crea líneas de igual fase que forman la espiral observada.

- **Dinámica inducida por STT**:  
  El torque de transferencia de espín, especialmente con $\xi \neq 0$, genera una **deriva** del núcleo y distorsiona las líneas de fase.  
  Esto provoca que las espirales se estiren o se deformen a medida que avanza el tiempo.

#### Conexión con la ecuación de LLG:
En el plano \(xy\), la parte precesional de la ecuación LLG:
$
\frac{\partial \mathbf{M}_\perp}{\partial t} 
= -\gamma \, \mathbf{M} \times \mathbf{H}_\text{ef} 
+ \xi \, \mathbf{M} \times \left[ (\mathbf{u} \cdot \nabla) \mathbf{M} \right]
$
hace que la magnetización gire alrededor del núcleo, mientras que el término no adiabático ajusta la velocidad de relajación hacia el patrón espiral estable.

#### Conclusión:
- Las espirales son la **huella directa** del estado tipo vórtice en el plano.
- La forma y el sentido de giro de las espirales indican el **quiroalidad** del vórtice (horario o antihorario).
- Cambios en $\xi$ no alteran la topología básica de la espiral, pero sí su **rapidez de estabilización** y el **grado de deformación** durante la dinámica.

## Comparación de estados iniciales y finales para $M_x$, $M_y $ y $ M_z$

A continuación, se presentan las configuraciones de magnetización **iniciales** (t ≈ 0 ns) y **finales** (t = 15 ns) para las tres componentes de la magnetización$ (M_x, M_y, M_z)$ en distintos valores de la constante no adiabática $\xi$.

El estado inicial en estas simulaciones corresponde a un **vórtice magnético**.  
En este estado, la magnetización está predominantemente en el plano (XY) y gira alrededor del centro:

- El núcleo del vórtice apunta fuera del plano $(M_z \approx \pm 1)$.
- Fuera del núcleo, la magnetización forma un patrón circular, con componentes \(M_x\) y \(M_y\) que cambian de signo dependiendo de la posición angular.

En coordenadas cartesianas:

- Para \(M_x\):  
  - Lado derecho: magnetización hacia \(+x\) → **rojo**.  
  - Lado izquierdo: magnetización hacia \(-x\) → **azul**.
- Para \(M_y\):  
  - Parte superior: magnetización hacia \(+y\) → **rojo**.  
  - Parte inferior: magnetización hacia \(-y\) → **azul**.

Ese patrón azul–rojo es la **proyección de la textura circular** sobre un eje fijo y representa un equilibrio inicial estable del vórtice.

### Evolución bajo corriente de espín

Cuando se aplica la corriente:

1. **Torque adiabático $(\xi = 0)$**:  
   - Desplaza el vórtice sin destruir instantáneamente el patrón circular.  
   - La transición rojo–azul se mantiene más tiempo.

2. **Torque no adiabático $(\xi > 0)$**:  
   - Introduce un empuje perpendicular que rompe la simetría.  
   - Acelera la deformación del patrón y la pérdida del estado de vórtice.
   - Con valores altos (\(\xi \approx 0.5\)), el patrón desaparece rápidamente, conduciendo a un estado más uniforme.

En los primeros instantes (\(< 0.4\ \text{ns}\)) todavía se observa un **semi-equilibrio** azul–rojo, pero este es transitorio y se pierde al aumentar la influencia del torque no adiabático y del damping.


### 1. Evolución general

- **Estado inicial (t ≈ 0 ns)**:  
  - El sistema parte de una configuración tipo vórtice o de simetría definida según la condición inicial.
  - Las tres componentes presentan una distribución suave, con transiciones graduales entre regiones de diferente polarización.
  - La energía está distribuida entre precesión y deformaciones iniciales del vórtice.

- **Estado final (t = 15 ns)**:  
  - La dinámica del torque de transferencia de espín (STT) y el amortiguamiento $ \alpha$ llevan al sistema a un estado estable.
  - Las regiones de magnetización bien definidas reemplazan gradualmente a la distribución inicial.
  - La constante $\xi $controla la rapidez y nitidez con la que se alcanza este patrón.

---

### 2. Componente \(M_x\)

- **Inicial**:  
  Patrón simétrico con regiones de polarización opuesta (rojo y azul) que se extienden de forma continua.  
  Gradientes suaves indican que el sistema aún no ha formado dominios rígidos.

- **Final**:  
  Dos dominios claramente separados con fronteras bien definidas.  
  En $\xi$ altas, estas fronteras se forman antes y son más abruptas.

---

### 3. Componente \(M_y\)

- **Inicial**:  
  Dominio con simetría transversal, pero sin una frontera abrupta.  
  La transición rojo-azul es más gradual debido a la ausencia de distorsiones fuertes.

- **Final**:  
  División clara entre polarización positiva y negativa en $y.  $
  Para $\xi $ grandes, el transporte transversal es más eficiente, acelerando la definición de estas regiones.

---

### 4. Componente \(M_z\)

- **Inicial**:  
  Núcleo fuera del plano concentrado, con el resto de la muestra en polarización opuesta suave.  
  El contraste es bajo debido a que el sistema aún está cerca de su condición inicial.

- **Final**:  
  Núcleo bien definido con alta polarización fuera del plano rodeado de una región opuesta.  
  La forma y el tamaño del núcleo son similares para todas las \(\xi\), pero su estabilización es más rápida para valores altos.

---

### 5. Interpretación física

La dinámica está descrita por la ecuación LLG con STT:

$
\frac{\partial \mathbf{M}}{\partial t} = 
- \gamma \, \mathbf{M} \times \mathbf{H}_\text{ef} 
+ \frac{\alpha}{M_s} \, \mathbf{M} \times \frac{\partial \mathbf{M}}{\partial t}
- (\mathbf{u} \cdot \nabla) \mathbf{M}
+ \xi \, \mathbf{M} \times \left[ (\mathbf{u} \cdot \nabla) \mathbf{M} \right]
$

- El **término adiabático** transporta la textura magnética a lo largo de la dirección de la corriente.
- El **término no adiabático** (\(\xi\)) introduce un torque transversal que acelera la relajación.
- El patrón final depende más de la geometría y condiciones iniciales que de la magnitud exacta de \(\xi\), pero \(\xi\) modula el tiempo de transición y la definición de las fronteras de dominio.

---

### Conclusión:
- Independientemente del valor de $\xi$, el sistema converge a configuraciones de dominio similares debido al equilibrio entre precesión, amortiguamiento y STT.
- La constante $\xi$ no cambia drásticamente la topología final, pero sí controla la **velocidad de transición** y la **nitidez de las fronteras de dominio**.
- Este comportamiento es coherente con la física del STT: en ausencia de anisotropías adicionales, el sistema tiende a estados de mínima energía que dependen más de la geometría y condiciones iniciales que de la magnitud exacta de $\xi$.

````python
import numpy as np
import matplotlib.pyplot as plt
import glob

# Lista de valores de xi
adiabaticc = [0, 0.05, 0.1, 0.5]
slice_index = 0  # Corte en z

for xi in adiabaticc:
    sim_name = f"SP5_xi_{xi}"
    npy_files = sorted(glob.glob(f"{sim_name}.out/m*.npy"))
    
    if not npy_files:
        print(f"No se encontraron archivos .npy para xi={xi}")
        continue
    
    # Tomar el último frame
    m = np.load(npy_files[-1])
    
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    componentes = ['M_x', 'M_y', 'M_z']
    cmaps = ['RdBu', 'RdBu', 'RdBu']
    
    for i, ax in enumerate(axes):
        slice_data = m[i, slice_index, :, :]
        pcm = ax.pcolormesh(slice_data, cmap=cmaps[i], vmin=-1, vmax=1, shading='auto')
        ax.set_title(f"{componentes[i]} final (ξ={xi})")
        ax.set_xticks([])
        ax.set_yticks([])
        fig.colorbar(pcm, ax=ax, label=componentes[i])
    
    plt.tight_layout()
    plt.show()
````
